a) Crea un mapa que contenga los datos del archivo paises.csv en el año 2013
    Clave: El nombre del país
    Valor: La expectativa de vida

```python
file=open("paises.csv")
lector = csv.reader(file)
lista1=[x for x in lector if x[2]=='2013']
mapa2013 = mapa()
for elemento in lista1:
    pais=elemento[0]
    expectativa=float(elemento[1])
    mapa2013[pais]=expectativa
for b in mapa2013:
    print("°Clave:",b,"°Valor:" ,mapa2013[b])
```

c) Encuentra la expectativa de vida para cada uno de los paises anteriores.
    (supongo que aqui es sin especificar el año)

f) Calcula el promedio de expectativa de vida para todos los paises que empiezan
con C.

```python
mapaC=mapa()
paisesC=[]
for elemento in mapaF:
    if elemento[0]=='C':
        paisesC.append(elemento)
file=open("paises.csv")
lector = csv.reader(file)
listaC=[x for x in lector if x[0] in paisesC]
for datos in listaC:
    pais=datos[0]
    mapaC[pais]+=float(datos[1])
den=[]
for x in paisesC:
    cont=0
    for d in listaC:
        if x==d[0]:
            cont+=1
    den.append(cont)
i=0
for w in mapaC:
    mapaC[w]=mapaC[w]/den[i]
    i+=1
    print(w, "-Promedio de exp. de vida: ", mapaC[w])
```
